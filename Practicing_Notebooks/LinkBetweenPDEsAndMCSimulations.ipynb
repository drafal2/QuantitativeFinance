**We want to connect a solution to Black-Scholes-Merton PDE derived from arbitrage-free pricing (construction of replicating portfolio) with Monte Carlo paths generation of risk-neutral GBM stochastic differential equation.**

So we have a market described by two equations:

$dM(t) = rM(t)dt$ <br>
$dS(t) = \mu S(t)dt + \sigma S(t)dW^\mathbb{P}(t)$

and contingent claim (conditional payoff depanding on behaviour of underlying) of the form:

$\chi = V(S(T), T)$

Arbitrage-free price is given, via Ito's Lemma, by $V(S, t)$, where function V(S, t) satisfies the Black-Scholes equation:

$\frac{\partial V}{\partial t} + rS \frac{\partial V}{\partial S} + \frac{1}{2}\sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} - rV = 0$ <br>
$V(S, T) = \chi$


#### **Feynman-Kac formula**

It offers a link between PDEs and stochastic process. If offers a method for solving PDEs by simulating random paths of a stochastic process.

Suppose we are given the PDE:<br>
$\frac{\partial V}{\partial t} + \overline{\mu} (x, t) \frac{\partial V}{\partial x} + \frac{1}{2}\overline{\sigma}^2 (x, t)) \frac{\partial^2 V}{\partial x^2} = 0$<br>

subject to a final condition:<br> 
$V(x, t) = \eta(x)$

then Feynman-Kac formula suggests: <br>
$V(x, t) = \mathbb{E}(\eta(X(T)) | \mathcal{F(t)})$

Where $X(t)$ is an Ito process driven by the equation $dX(t) = \overline{\mu}(X(t), t)dt + \overline{\sigma}(X(t), t)dW(t)$ with $W(t)$ is a Wiener process with initial for $X(t) = x$



**Application of option price formula**

$V(S, t)$ is the unique solution of the final condition problem:

$\frac{\partial V}{\partial t} + rS \frac{\partial V}{\partial S} + \frac{1}{2}\sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} - rV = 0$ <br>
$V(S, T)$ is given

The solution can be obtained as:

$V(S, t) = e^{-r(T-t)} \mathbb{E}^{\mathbb{Q}}[V(S(T), T) | S(t)]$<br>
given that $dS(t) = rS(t)dt + \sigma S(t)dW^{\mathbb{Q}}(t)$

**To sum up, the Feynman-Kac formula says that when we have a payment on a certain time $T$, the value of a derivative in any time $t$ before $T$ is equal to expectations at time $t$ of a payoff from this derivative at the $T$**. It shows that the solution to certain PDEs can be written as an expectation (average) of a stochastic process. Of course, this value needs to be discounted until $t$ to get price of a derivative as of $t$. Expections can be calculated by Monte-Carlo paths simulation of underlying SDE.
