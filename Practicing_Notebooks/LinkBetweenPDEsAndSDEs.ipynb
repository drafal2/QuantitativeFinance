**Stochastic differential equation** is a differential equation in which one or more of the terms is a stochastic process, resulting in a solution which is also a stochastic process. It is mathematical equations describing systems influenced by random factors, adding a "noise" term (like Brownian motion) to traditional differential equations, resulting in solutions that are themselves random processes. Instead of a single solution, SDEs yield a stochastic process, a collection of possible paths or trajectories. Mathematical tools like the Itô calculus (and Itô's Lemma) are used to analyze and solve SDEs, as standard calculus rules don't directly apply due to the random noise. It has random component.

**Partial differential equation** is an equation which involves a multivariable function and one or more of its partial derivatives. It is mathematical equations involving an unknown function of multiple variables (like space and time) and its partial derivatives, used to model complex systems where changes depend on several factors, such as heat flow, fluid dynamics, waves, electricity, and quantum mechanics. They describe how quantities change in relation to these multiple dimensions, providing the mathematical language for many physical laws. Unlike ordinary differential equations (ODEs) for single-variable systems, PDEs handle systems spread across multiple dimensions (space and time). A PDE sets up a relationship between the function and its rates of change (derivatives). Solving the PDE means finding the specific unknown function that satisfies that relationship, often revealing the system's behavior over time and space. Solutions can be found analytically (exact formulas) or numerically (using computer models for complex problems). It does not have random component.

**The link between SDEs and PDEs**

The link between Stochastic Differential Equations (SDEs) and Partial Differential Equations (PDEs) is one of the most powerful results in modern mathematics, specifically through the Feynman-Kac formula. This connection allows us to solve difficult PDEs by simulating random paths (Monte Carlo methods) or, conversely, to find the expected values of stochastic processes using deterministic PDE solvers.

The core link is **Feynman-Kac formula**. It is the idea that the solution to a parabolic PDE can be represented as the conditional expectation of a stochastic process. 

When we have a stochastic process like GBM of a stock price $S_t$: $dS_t = \mu S_tdt + \sigma S_tdW_t$, we want to find the price of an option which is the expected payoff $V(S, t) = \mathbb{E}[e^{(-r(T-t)}Payoff(S_T) | S_t=S]$. We can find the solution in two ways:
1. **The SDE**: we simulate the individual paths of a stock price and calculate expected payoff
2. **The PDE**: we can derive the Black-Scholes PDE and solve it as it describes how expectations (average) of all possible paths evolve over time

We can derive PDE from SDE using link given by Feynman-Kac theorem:

Suppose that variable $S_T$ follows SDE: <br>
$dX_t = \mu(X_t, t)dt + \sigma(X_t, t)dW(t)$

You want to find the value of a function $u(x, t)$ = $\mathbb{E}[g(X_T) | X_t=x]$, the Feynman-Kac formula states that $u(x, t)$ satisfies PDE: <br>
$\frac{\partial u}{\partial t} + \mu (x, t) \frac{\partial u}{\partial x} + \frac{1}{2}\sigma^2 (x, t) \frac{\partial^2 u}{\partial x^2} = 0$<br>



#### **The role of Ito's lemma**

1. **The Mechanics (Ito’s Lemma)**
Ito’s Lemma describes the dynamics of a function $f(X_t, t)$. It breaks the change into a deterministic drift ($dt$) and a stochastic noise ($dW_t$). This is the "raw" SDE form.

2. **The Transition to PDE**
To move from a random SDE to a deterministic PDE, the "random noise" must be removed. This happens in two different ways depending on your goal:
- The Statistical Path (Expectation): By taking the expectation ($E$), the random $dW_t$ term mathematically vanishes because its average is zero. This tells you the "average" outcome of the process.
    - Result: The Kolmogorov Forward/Backward equations.
- The Financial Path (Hedging): In Black-Scholes, we don't just "hope" for the average; we cancel the noise. By creating a portfolio of the option and the underlying stock (Delta Hedging), the $dW_t$ terms of both assets subtract to zero.
    - Result: The Black-Scholes PDE.
    
3. **The Unifying Link (Feynman-Kac)**
The ultimate bridge is the Feynman-Kac Theorem. It proves that the solution to a PDE can be written as an expectation.
- The Key Insight: Even though Black-Scholes uses hedging to "eliminate" noise, the resulting PDE is identical to one solved by taking an expectation - provided you replace the real-world growth ($\mu$) with the risk-free rate ($r$). This is why we can solve PDEs using Monte Carlo simulations.

**Ito's Lemma provides the dynamics. The PDE is formed by eliminating the $dW_t$ term - either through averaging (Expectation) or offsetting (Hedging). Once the noise is gone, the PDE can be solved numerically or via stochastic simulation (Monte Carlo)**
