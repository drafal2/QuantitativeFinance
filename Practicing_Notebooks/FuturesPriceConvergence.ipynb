```python
url = "https://stooq.com/q/d/l/?s=fw20z2520&d1=20241223&d2=20251219&i=d"

try:
    # 2. Download and read the CSV
    # Note: Stooq often requires a User-Agent header to allow the download
    df_futures = pd.read_csv(url)

    # 3. Data Cleaning
    if not df_futures.empty:
        # Convert Date column to datetime objects
        df_futures['Date'] = pd.to_datetime(df_futures['Date'])
        df_futures.set_index('Date', inplace=True)
        
        print("Data successfully downloaded!")
        print(df_futures.head()) # Show the most recent rows
    else:
        print("The downloaded file is empty.")

except Exception as e:
    print(f"Error: {e}")
```
